```python
import numpy as np
from PIL import Image

IMG_SIZE = 224

def predict_pneumonia(img):
    img = img.resize((IMG_SIZE, IMG_SIZE))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prob = model.predict(img_array)[0][0]

    if prob > 0.5:
        label = "PNEUMONIA"
        color = "#ff4c4c"   # RED
    else:
        label = "NORMAL"
        color = "#00c851"   # GREEN

    # ‚úÖ HTML styled output
    result_html = f"""
    <div style="
        padding: 20px;
        border-radius: 12px;
        background-color: {color};
        color: white;
        font-size: 20px;
        font-weight: bold;
        text-align: center;">
        ‚úÖ Prediction: {label} <br>
        üìä Confidence: {prob:.4f}
    </div>
    """

    return result_html
```

```python
import gradio as gr
import numpy as np
from PIL import Image
import tensorflow as tf

# ‚úÖ Load your trained model
model = tf.keras.models.load_model("/content/drive/MyDrive/pneumonia_final.keras")

IMG_SIZE = 224

# ‚úÖ Prediction function with colored HTML output
def predict_pneumonia(img):
    img = img.resize((IMG_SIZE, IMG_SIZE))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prob = model.predict(img_array)[0][0]

    if prob > 0.5:
        label = "PNEUMONIA"
        color = "#ff4c4c"   # RED
    else:
        label = "NORMAL"
        color = "#00c851"   # GREEN

    result_html = f"""
    <div style="
        padding: 20px;
        border-radius: 12px;
        background-color: {color};
        color: white;
        font-size: 20px;
        font-weight: bold;
        text-align: center;">
        ‚úÖ Prediction: {label}<br>
        üìä Confidence: {prob:.4f}
    </div>
    """

    return result_html


# ‚úÖ UI with HTML output (THIS IS THE FIX)
ui = gr.Interface(
    fn=predict_pneumonia,
    inputs=gr.Image(type="pil", label="Upload Chest X-ray"),
    outputs=gr.HTML(label="Prediction Result"),  # ‚úÖ MUST be gr.HTML
    title="ü´Å Pneumonia Detection from Chest X-ray",
    description="Upload a chest X-ray image to detect Pneumonia using a Deep Learning (MobileNetV2) model.",
    allow_flagging="never"
)

ui.launch()
```
